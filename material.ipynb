```python
# Create a Faker instance with a specific locale
fake = Faker('yo_NG')  # Yoruba (Nigeria)
```

```python
materials = []
```

```python
MATERIAL_NAMES = [
    {"Fluids" :
    ["Brake Fluid DOT 3",
    "Brake Fluid DOT 4",
    "Engine Oil 10W-30",
    "Engine Oil 15W-40",
    "Engine Oil 20W-50",
    "Gear Oil 80W-90",
    "Hydraulic Oil ISO 32",
    "Hydraulic Oil ISO 46",
    "Coolant Concentrate",
    "Antifreeze Solution",
    "Windshield Washer Fluid",
    "Power Steering Fluid"], 
    
    "Greases" :
    ["Lithium Grease EP2",
    "Lithium Grease EP3",
    "Multi-purpose Grease",
    "High Temperature Grease",
    "Bearing Grease",
    "Chassis Grease"],
    
    "Paints & Coatings" :
    ["Anti-rust Primer Grey",
    "Anti-rust Primer Red Oxide",
    "Enamel Paint Black",
    "Enamel Paint White",
    "Spray Paint Silver",
    "Epoxy Paint",
    "Zinc Primer",
    "Anti-corrosion Coating"],
    
    "Tyres" :
    ["Tyre 145/80 R12",
    "Tyre 165/80 R14",
    "Tyre 175/65 R15",
    "Tyre 185/70 R14"],
    
    "Filters" :
    ["Oil Filter",
    "Air Filter",
    "Fuel Filter",
    "Cabin Air Filter"]
    }
]
```

```python
# Flatten the list of dictionaries
ALL_MATERIALS = []
for dict_item in MATERIAL_NAMES:
    for category, items in dict_item.items():
        for item in items:
            ALL_MATERIALS.append((category, item))

# Now pick randomly
```

```python
UOM_BY_CATEGORY = {
    'Fluids': random.choice(['L', 'Ltr', 'Liters']),  # Add inconsistencies
    'Greases': random.choice(['KG', 'Kg', 'kg']),
    'Paints': random.choice(['L', 'Ltr']),
    'Tyres': 'EACH',
    'Filters': 'EACH',
    'Fasteners': random.choice(['PCS', 'Pcs', 'pieces']),
}
```

```python
for i, (category, item_name) in enumerate(ALL_MATERIALS):

    material = {
        'number' : f'MAT{i+1:05d}',
        'item_name' : item_name,
        'item_type' : category,
        'unit_of_measure' : UOM_BY_CATEGORY.get(category, 'EACH'),
        'standard_price' : f'{random.uniform(500, 100000):.0f}',
        'current_stock_quantity' : random.randint(0, 1000),
        'last_issue_date' : fake.date_between(start_date='-1y', end_date='-0y')
    }

    materials.append(material)

```

```python
# Now you have 500 materials stored
print(f"Generated {len(materials)} materials")
print(materials[0])  # Print first one as example
```

```python
# Create datasets and save to CSV
df_materials = pd.DataFrame(materials)
df_materials.to_csv('materials.csv', index=False)
```
